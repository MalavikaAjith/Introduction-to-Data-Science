```python
import numpy as np
import pandas as pd
```

```python
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.groupby('Company')
a = df.groupby('Company')
```

```python
a.mean()
```

```python
a.std()
```

```python
a.min()
```

```python
a.max()
```

```python
a.count()
```

```python
a.describe()
```

```python
a.describe().transpose()
```

```python
a.describe().transpose()['GOOG']
```
